# Open CV 

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("fury.png")
img
```

```python
plt.imshow(img)
plt.show()              

#how the hell did our banner turn blue XD ???
#It's because open cv uses BGR format by default
#and matplotlib uses RGB
```

```python
#let's change it to RGB:

original_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(original_img)
plt.show()
```
