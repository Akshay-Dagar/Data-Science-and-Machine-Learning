```python
import numpy as np
import pandas as pd
```

```python
x_train=pd.read_csv("Logistic_X_Train.csv").values
y_train=pd.read_csv("Logistic_Y_Train.csv").values.flatten()
x_test=pd.read_csv("Logistic_X_Test.csv").values

x_train.shape,y_train.shape,x_test.shape
```

```python
col_ones=np.ones((x_train.shape[0],1))
x_train=np.hstack((x_train,col_ones))
x_train.shape
```

```python
col_ones=np.ones((x_test.shape[0],1))
x_test=np.hstack((x_test,col_ones))
x_test.shape
```

```python
x_train=(x_train-np.mean(x_train,axis=0))/(np.std(x_train,axis=0))
x_test=(x_test-np.mean(x_test,axis=0))/(np.std(x_test,axis=0))

x_train[:,-1]=1
x_test[:,-1]=1
x_test
```

```python
class LogisticRegression():
    
    def __init__(self,lr,itr=100,stop=0.00001):
        
        self.lr=lr
        self.itr=itr
        self.stop=stop
        
    def sigmoid(self,x):
        return 1/(1+(np.e**(-np.dot(x,self.theta))))
    
    def gradient(self,x,y):
        
        hyp=self.sigmoid(x)
        
        return -(1/x.shape[0])*(((hyp-y).reshape(-1,1)*x).sum(axis=0))
    
    def likelihood(self,x,y):

        return (1/x.shape[0])*(np.sum((y*np.log2(self.sigmoid(x)))+((1-y)*np.log2(1-self.sigmoid(x)))))
        
    def fit(self,x,y):                               #x and y are training points and labels
        
        self.theta=np.zeros(x.shape[1])      #intializing parameters
        
        prev_like=0
        for i in range(self.itr):
            
            grad=self.gradient(x,y)
            
            self.theta=self.theta+(self.lr*grad)
            
            likelihood=self.likelihood(x,y)
            print(likelihood)
            
            #if abs(likelihood-prev_like)<self.stop:
            #    break
                
            prev_like=likelihood
            
    def predict(self,x):
        
        return np.round(self.sigmoid(x))                  #if hyp>=0.5, it will return 1 else 0
    
        
```

```python
clf=LogisticRegression(10,1000)
clf
```

```python
clf.fit(x_train,y_train)
```

```python
predictions=np.int32(clf.predict(x_test))
predictions
```

```python
d={"label":predictions}
df=pd.DataFrame(d)

df.to_csv("submission.csv",index=False)
```
