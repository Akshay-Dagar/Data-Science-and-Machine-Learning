# Bar graphs

```python
x=[1,2,3]          #bars will be centred around these points
y=[10,20,40]       #heights of the bars

plt.bar(x,y,color="orange")
plt.show()
```

#### Suppose we want to compare 2 graphs, for example, prices of items for first year vs second year:

```python
x1=np.array([2,4,6])          
y1=np.array([10,20,40])
x2=x1+0.5
y2=[5,30,20]

plt.bar(x1,y1,color="orange",width=0.5,label="year1",tick_label=["Robb","Stannis","Tommen"])
plt.bar(x2,y2,color="blue",width=0.5,label="year2")

#width is specified so that 2 adjacent bars 
#don't overlap (try removing width and see), 
#be default, width is 1
#tick label specifies the x label for each bar

plt.xlim(1,7)               #x units will be from 1 to 7   (our graph was going on negative x, so this is an easy fix)
plt.ylim(0,45)              #y units will be from 0 to 45   

plt.legend()                #to show the labels
plt.show()
```

## Other probability distributions:

#### Uniform Distribution:

###### All data points have same probability. This is not generally the case in real life. Mostly, normal distribution is the most popular.

#### Multivariate Normal Distribution: 
###### As the name suggests, this is a normal distribution involving multiple variables, the variables are correlated according their covariance. For example:

###### See that here x and y have a very high covariance (they are highly correlated, as x increases, y increases and vice versa). Whereas the following distribution on the other hand has verly less covariance:

###### Multivariate Normal distr. is defined by mean,std dev,covaraince:
###### mean: Data is centred around [meanx,meany]
###### std_dev: Most of the data points lie within [devx,devy] from mean
###### 
###### see that std_dev_y is less than std_dev_x. Also, see that convariance between x and y in this dataset is less (they have very less correlation)
###### Co-Variance: This is a matrix defined as:


###### sigmaXX is simply the std_dev_x, sigmaYY is simply the std_dev_y. sigmaXY denotes the correlation(covariance) between x and y). sigmaYX==sigmaXY
###### sigmaXY=1 would mean that data is along y=x line

## Let's generate a Multivariate Normal Distribution using NUMPY:

```python
mean=np.array([0,0])                #data is centred around origin
covariance=np.array([[1,0],
                     [0,1]])

#sigmaXX and sigmaYY are both 1, and sigmaXY and
#sigmaYX ae both 0 meaning no correlation bet. x,y

distr=np.random.multivariate_normal(mean,covariance,500)
distr
#see that distr is a 2d array, 0th column represent x and 1st column represents y
```

```python
x=distr[:,0]
y=distr[:,1]

plt.scatter(x,y,color="yellow")
plt.show()                        
#see that covariance is non-existant and mean is [0,0] and 
#most point lie within 1 (along x) and 1 (along y) from mean
```

```python
#Let's increase the covaraince to 0.8 and see:
mean=np.array([0,0])                
covariance=np.array([[1,0.8],
                     [0.8,1]])
distr=np.random.multivariate_normal(mean,covariance,500)
x=distr[:,0]
y=distr[:,1]
plt.scatter(x,y,color="yellow")
plt.show()                           #see that covariance is high now
```

```python
#Let's increase the covaraince to 1.0 and see:
mean=np.array([0,0])                
covariance=np.array([[1,1],
                     [1,1]])
distr=np.random.multivariate_normal(mean,covariance,500)
x=distr[:,0]
y=distr[:,1]
plt.scatter(x,y,color="yellow")
plt.show()                           #see that covariance is high now
```

```python
#let's change std_dev now:
#Let's increase the covaraince to 1.0 and see:
mean=np.array([0,0])                
covariance=np.array([[1.2,0.9],
                     [0.9,1.5]])
distr=np.random.multivariate_normal(mean,covariance,500)
x=distr[:,0]
y=distr[:,1]
plt.scatter(x,y,color="yellow")
plt.show()                           #see that covariance is high now
```
