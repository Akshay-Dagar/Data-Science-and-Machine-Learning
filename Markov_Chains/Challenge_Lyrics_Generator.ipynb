```python
import numpy as np
```

```python
def read_txt(filepath):
    
    with open(filepath,"r",encoding="utf8") as f:
        return f.read()
```

```python
txt=read_txt("kolaveri.txt")
txt=txt.lower()
txt
```

```python
def build_table(text,k=4):                     #k is the no. of words to be used as x to predict y (the next character)
    
    trans={}                                       #transition table (dict)
    
    #get frequencies for each word,prediction char pair:
    
    for i in range(len(text)-k):                        #go till k th character from last
        
        word=text[i:i+k]
        char=text[i+k]
        
        if trans.get(word) is None:            #if word is not present in transition table
            trans[word]={}                     #insert the word into trans dict and make it's value as an empty dict
            trans[word][char]=1                #insert char and make it's value as 1 in word's dict in trans
            
        else:
            if trans[word].get(char) is None:
                trans[word][char]=1
            else:
                trans[word][char]+=1
    
    #convert the frequencies into probabilities:
    
    for w in trans.keys():
        sum=np.sum(list((trans[w].values())))          #sum of all frequencies corresponding to word w
        
        for c in trans[w].keys():
            trans[w][c]=trans[w][c]/sum
            
    return trans
```

```python
table=build_table(txt,k=4)
table
```

```python
def generate_txt(start,length,k=4):    
    
    np.random.seed(11)                                            
    #just for submission to coding blocks (so that same lyrics are generated always 
    #(because np.random.choice() makes same choice always because random seed is set))
    
    '''start should have at least K characters, 
       length specifies no. of chars to be generated'''
    
    sentence=start
    
    word=start[-k:]
    for i in range(length):
        
        if table.get(word) is None:
            sentence+=' '
        else:
            choices=list(table[word].keys())
            prob=list(table[word].values())
            
            sentence+=np.random.choice(choices,p=prob)
            
        word=sentence[-k:]
        
    return sentence
```

```python
sub=generate_txt("yo boys",2000)
print(sub)
```
