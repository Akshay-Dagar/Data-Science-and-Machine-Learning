```python
#importing modules
from pandas import *
from numpy import *
from sklearn import datasets
from collections import Counter
from math import log
```

```python
def entropy(y):
    count=Counter(y)
    #formula: -sum(pi*log(pi))
    sum=0
    for i in count.keys():
        pi=count[i]/y.shape[0]       #probability of class i
        sum+=(pi*math.log(pi,2))
    return (-sum)
```

```python
def printnodedata(y,level):
    print("Level ",level)
    count=Counter(y)    #count is a dictionary, with keys as flower type(0,1,2) and values as no. of times that flower is in y
    for i in count.keys():
        print("Count of",i,"=",count[i])
    print("Current Entropy=",entropy(y))
```

```python
def info_gain(x,y,ftr):
    #formula=entropy_initial-entropy_final
    entropy_i=entropy(y)
    s=set(x[:,ftr])
    entropy_f=0
    for i in s:
        y_i=y[x[:,ftr]==i]                                 
        entropy_f+=(y_i.shape[0]/y.shape[0])*entropy(y_i)
    return entropy_i-entropy_f
```

```python
def printtree(x,y,features,level,deleted_features):
    
    #printing data about current node:
    printnodedata(y,level)
    
    #base cases:
    if len(set(y))==1:                                              #node is pure
        print("Reached a leaf Node")
        print()
        return
    if len(deleted_features)==features.shape[0]:                                        #no more features left to split on
        print("Reached a leaf Node")
        print()
        return
    print("Splitting on feature",end=" ")
    
    #splitting based on a feature:
    max_gr=float('-inf')                                            #max gain ratio on split(initializing
    max_f=-1                                                        #feature that gives max gain ratio
    for i in range(len(features)):
        if i in deleted_features:
            continue
        gr=info_gain(x,y,i)                     #calculating gain ratio for i'th feature 
        if gr>max_gr:
            max_gr=gr
            max_f=i
    print(features[max_f],"with gain ratio:",max_gr)
    print()
    deleted_features.append(max_f)                                #deleting split feature from the list of all features
    for i in set(x[:,max_f]):
        xf=x[x[:,max_f]==i]
        yf=y[x[:,max_f]==i]
        printtree(xf,yf,features,level+1,deleted_features)
```

```python
#importing iris dataset and spltting into x_train and y_train:

iris=datasets.load_iris()
df=DataFrame(iris.data,columns=iris.feature_names)
y=iris.target                                               #0=>Iris-Setosa, 1=>Iris-Versicolor, 2=>Iris-Virginica
features=array(df.columns)
deleted_features=[]
x=df.values
```

```python
printtree(x,y,features,0,deleted_features)
```
