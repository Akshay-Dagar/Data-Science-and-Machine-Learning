### Creating Dataset:

```python
import requests

english_most_common_1k = 'https://gist.githubusercontent.com/deekayen/4148741/raw/98d35708fa344717d8eee15d11987de6c8e26d7d/1-1000.txt'

#Get the file of 1K most common words from TXT file in a github repo
response = requests.get(english_most_common_1k)
common = set(response.text.split('\n'))
print(common)
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize,RegexpTokenizer
from textstat.textstat import textstatistics
```

```python
def AverageSentenceLength(text,words):
    
    n_sentences=len(sent_tokenize(text))
    n_words=len(words)
    return (n_words/n_sentences)
```

```python
def PercentageDifficultWords(words):
    """Percentage of words that are uncommon (not in top 10k most common english words."""
    
    n_difficult=0
    
    for word in words:
        if word not in common:
            n_difficult+=1
            
    return n_difficult/len(words)
```

```python
def AverageWordLengthInSyllables(words):
    
    n_syllables=0
    
    for word in words:
        n_syllables+=textstatistics().syllable_count(word)
        
    return n_syllables/len(words)
```

```python
def CreateFeatureSet(text):
    
    tokenizer=RegexpTokenizer("[a-zA-Z]+")
    words=tokenizer.tokenize(text)
    
    return [AverageSentenceLength(text,words), PercentageDifficultWords(words), AverageWordLengthInSyllables(words)]
```

```python
x_train=[]
y_train=train_data.target.values

for excerpt in tqdm_notebook(train_data.excerpt.values):
    x_train.append(CreateFeatureSet(excerpt))
    
x_train=np.array(x_train)
x_train
```

```python
x_test=[]
test_ids=test_data.id.values

for excerpt in test_data.excerpt.values:
    x_test.append(CreateFeatureSet(excerpt))
    
x_test=np.array(x_test)
x_test
```

### Random Forest Model:

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=100,max_depth=9,min_samples_split=5,min_samples_leaf=5)
```

```python
reg.fit(x_train,y_train)
```

```python
train_predictions=reg.predict(x_train)
val_predictions=reg.predict(x_val)
```

```python
from sklearn.metrics import mean_squared_error
val_error=mean_squared_error(y_val,val_predictions)
train_error=mean_squared_error(y_train,train_predictions)
train_error,val_error
```
